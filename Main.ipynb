```python
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
iris_df = pd.DataFrame(data=iris.data , columns=iris.feature_names)
iris_df['species'] = iris.target


```

####Check whether it is null

```python
iris_df.isnull().sum()
```

###Data pre process

```python
# prompt: write the full code for dataset for eda

import matplotlib.pyplot as plt
# Check for missing values
missing_values = iris_df.isnull().sum()

# Print the number of missing values in each column
print(missing_values)

# Get descriptive statistics of the data
description = iris_df.describe()

# Print the descriptive statistics
print(description)

# View the first 10 rows of the data
print(iris_df.head(10))

# Check the distribution of the target variable
iris_df['species'].value_counts().plot(kind='bar')
plt.show()

# Check the distribution of each feature
for col in iris_df.columns:
  if col != 'species':
    iris_df[col].hist()
    plt.show()

# Check the correlation between features
correlation_matrix = iris_df.corr()

# Print the correlation matrix
print(correlation_matrix)
```

```python
from sklearn.preprocessing import LabelEncoder , StandardScaler

# Label encoder the 'species' column
label_encoder = LabelEncoder()
iris_df['species'] = label_encoder.fit_transform(iris_df['species'])

scaler = StandardScaler()
iris_df[iris_df.columns[0:4]] = scaler.fit_transform(iris_df[iris_df.columns[0:4]])

iris_df
```
